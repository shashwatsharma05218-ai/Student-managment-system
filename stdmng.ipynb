# Student Management System

```python
import sqlite3
```

```python
conn = sqlite3.connect("student.db")

cursor = conn.cursor()

print("Database Connected Successfully")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS students(

id INTEGER PRIMARY KEY AUTOINCREMENT,

name TEXT,

age INTEGER,

gender TEXT,

course TEXT,

semester INTEGER,

email TEXT,

phone TEXT,

attendance REAL,

cgpa REAL

)
""")

conn.commit()

print("Table Created Successfully")
```

```python
def add_student():

    name = input("Enter Name : ")

    age = int(input("Enter Age : "))

    gender = input("Gender : ")

    course = input("Course : ")

    semester = int(input("Semester : "))

    email = input("Email : ")

    phone = input("Phone : ")

    attendance = float(input("Attendance : "))

    cgpa = float(input("CGPA : "))

    cursor.execute("""
    INSERT INTO students
    (name,age,gender,course,semester,email,phone,attendance,cgpa)

    VALUES (?,?,?,?,?,?,?,?,?)
    """,

    (name,age,gender,course,semester,email,phone,attendance,cgpa))

    conn.commit()

    print("Student Added Successfully")
```

```python
add_student()
```

```python
def view_students():

    cursor.execute("SELECT * FROM students")

    rows = cursor.fetchall()

    for row in rows:

        print(row)
```

```python
view_students()
```

```python
def search_student():

    name = input("Enter Student Name : ")

    cursor.execute(

        "SELECT * FROM students WHERE name=?",

        (name,)
    )

    result = cursor.fetchall()

    if result:

        for student in result:

            print(student)

    else:

        print("Student Not Found")
```

```python
search_student()
```

```python
def delete_student():

    sid = int(input("Enter Student ID : "))

    cursor.execute(

        "DELETE FROM students WHERE id=?",

        (sid,)
    )

    conn.commit()

    print("Student Deleted Successfully")
```

```python
delete_student()
```

```python
def update_student():

    sid = int(input("Enter Student ID : "))

    new_name = input("New Name : ")

    new_cgpa = float(input("New CGPA : "))

    cursor.execute("""

    UPDATE students

    SET name=?,

    cgpa=?

    WHERE id=?

    """,

    (new_name,new_cgpa,sid))

    conn.commit()

    print("Updated Successfully")
```

```python
update_student()
```

```python
while True:

    print("\n===== STUDENT MANAGEMENT SYSTEM =====")

    print("1. Add Student")

    print("2. View Students")

    print("3. Search Student")

    print("4. Update Student")

    print("5. Delete Student")

    print("6. Exit")

    choice = input("Enter Choice : ")

    if choice=="1":

        add_student()

    elif choice=="2":

        view_students()

    elif choice=="3":

        search_student()

    elif choice=="4":

        update_student()

    elif choice=="5":

        delete_student()

    elif choice=="6":

        print("Thank You")

        break

    else:

        print("Invalid Choice")
```

```python
conn.close()

print("Database Closed")
```
